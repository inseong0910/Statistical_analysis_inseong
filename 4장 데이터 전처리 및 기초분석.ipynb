```python
# 4. 조건을 만족하는 레코드 추출하기
# 조건 : 이탈여부가 1, 즉 이탈로 판단된 고객만을 추출

churn_customer = df[df.이탈여부==1]
churn_customer.head()
```

```python
# 5. 특정 필드만 추출하기
df_1 = df[['고객ID', '방문빈도']]

df_1.head()
```

```python
# 5. 데이터 분할하기
# 고객ID를 기준으로 ID 500 이하와 초과 두 고객군으로 분할

under_500 = df[df.고객ID<=500]
upper_500 = df[df.고객ID>500]

under_500.tail()
```

```python
# 6. 데이터 추가하기
# 앞서 분할했던 고객ID 1~500 데이터셋에 501~1000 데이터셋을 추가

df_join = under_500.append(upper_500, ignore_index=True)

df_join
```

```python
# 7. 데이터 병합하기
# df_1(고객ID, 방문빈도)와 df_2(고객ID, 총 매출액')를 병합

# 7-1. 필드 추출
df_1 = df[['고객ID', '방문빈도']]
df_2 = df[['고객ID', '총_매출액']]

# 7-2. 데이터 병합
df_merge = df_1.merge(df_2)

# 7-3. 데이터 출력
df_merge.head()
```

```python
# 8. 파생변수 생성 
import numpy as np

df['남녀'] = np.where(df.성별 == 0, '남자', '여자')
# 성별==0이면 남자로, 1이면 여자로 표기

df[['고객ID', '성별', '남녀']]
```

```python
# 9. 척도 변환
# 고객의 총 매출액을 기준으로 500만원 이상이면 '상'
# 300만원 이상~500만원 미만이면 '중', 300만원 미만이면 '하'로 표현하는 고객등급 변수를 생성

# 9-1. 척도 변환 파생변수 생성
df['New고객등급'] = np.where(df.총_매출액>=5000000, '상', np.where(df.총_매출액>=3000000, '중', '하'))

# 9-2. 데이터 확인
df[['고객ID', '총_매출액', 'New고객등급']]
```

```python
# 10. 모델링을 활용해 파생변수 생성
# 고객 방문빈도, 총 매출액 변수를 활용하여 '고객 1회 평균매출액' 파생변수를 생성

# 10-1. 척도 변환 파생변수 생성
df['New_1회_평균매출액'] = df['총_매출액']/df['방문빈도']

# 10-2. 데이터 확인
df[['고객ID', '총_매출액', '방문빈도', 'New_1회_평균매출액']]
```

```python
# 11. 거래이력 요약을 통한 파생변수 생성
# 파생변수를 만드는 것은 아니지만, 거래이력을 요약하는 관점으로 새로운 데이터를 생성

# 11-1. 비 이탈고객 레코드 추출
churn_customer = df[df.이탈여부==1]
Non_churn_customer = df[df.이탈여부==0]

# 11-2. 매출합계 출력
print(sum(churn_customer.총_매출액))
print(sum(Non_churn_customer.총_매출액))
```

```python
# 11-1. 모듈 및 데이터 탑재
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt 
df = pd.read_csv('Ashopping.csv',sep=',', encoding='CP949')

# 11-2. 한글깨짐현상 방지 
matplotlib.rcParams['font.family'] = 'Malgun Gothic'   # '맑은 고딕'으로 설정 
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# 11-3. 히스토그램 작성하기
%matplotlib inline
plt.hist(df['서비스_만족도'], alpha=0.4, bins=7, rwidth=1, color ='red', label='서비스만족도')

# 11-4. 각종 옵션설정
plt.legend()
plt.grid()
plt.xlabel('서비스 만족도')
plt.ylabel('빈도')
plt.xticks(fontsize = 14) #글씨 크기설정
plt.yticks(fontsize = 14) #글씨 크기설정
plt.show()
```

```python
# 히스토그램 결과, 서비스만족도는 대부분 5~7점에 분포되어 있음
# 낮은 만족도를 나타낸 소수 고객들에 대한 불만족 원인을 파악하여 개선점을 찾을 필요가 있음
```

```python
# 12. 산점도 작성
# 산점도를 활용하여 방문빈도와 총매출액의 관계를 파악

%matplotlib inline
df.plot.scatter(x='방문빈도',y='총_매출액', grid=True, title='방문 빈도와 총 매출액간 관계')
```

```python
# 산점도 분석 결과, 방문빈도와 총매출액 사이에는 어느 정도 상관관계가 있는 것으로 파악되었음.
# 특히 누적 방문빈도 50회 전후에서는 방문빈도에 따라 매출액의 증가가 비교적 명확하였음.
# 50회 이상부터는 매출액의 편차가 매우 크게 나타남.
# 이 기업의 경우 우량고객이 되기 전까지는 매장방문을 유도하는 전략을,
# 우량고객이 된 이후에는 방문당 구매액을 증가시키는 전략을 수행하는 것이 더욱 바람직할 것임.
```

```python
# 13. 파이차트 작성
# 13-1. 구매유형별 고객 그룹화하기
groupby_구매유형 = df.groupby('구매유형')
groupby_구매유형.describe()

# 13-2. 파이차트 속성 지정
labels = ['1','2','3','4'] #라벨 지정
sizes = [43, 317, 144, 496] # 크기 지정
colors = ['yellowgreen','gold','lightskyblue','lightcoral'] #색 지정
explode = (0, 0.1, 0, 0)

# 13-3. 파이차트 작성하기
%matplotlib inline
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct='%1.1f%%',shadow = True, startangle = 90)
plt.axis('equal')
plt.title('구매유형에 따른 고객 분포')
plt.show()
```

```python
# 파이차트 결과, 전체 고객의 약 50%가 4번 그룹인 집중 구매형 고객군에 속하는 것을 알 수 있다.
# 1회성 구매형은 43명으로 가장 적은 고객들이 속해있다.
# 이 데이터는 VIP 고객들만 추출한 데이터이므로 1회성 구매고객이 오히려 적다.
# 각 고객 그룹별 특성에 맞는 마케팅을 기획하기 위해서는 각 고객 그룹에 대한 프로파일 분석과 같은 후속작업이 필요하다.
```
